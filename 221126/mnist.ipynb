```python
from mnist import load_mnist
import numpy as np
import pickle
```

```python
def get_data():
  (x_train, y_train), (x_test, y_test) = load_mnist(flatten=True, normalize=True, one_hot_label=False)
  return x_test, y_test
```

```python
def init_network():
  with open("sample_weight.pkl", 'rb') as f:
    network = pickle.load(f)
  return network
```

```python
def sigmoid(x):
  return 1 / (1+np.exp(-x))
```

```python
def softmax(x):
  m = np.max(x)
  exp_x = np.exp(x-m)
  exp_sum = np.sum(exp_x)
  
  return exp_x / exp_sum
```

```python
def predict(network, x):
  W1, W2, W3 = network['W1'], network['W2'], network['W3']
  b1, b2, b3 = network['b1'], network['b2'], network['b3']

  output = np.dot(x, W1) + b1
  output = sigmoid(output)

  output = np.dot(output, W2) + b2
  output = sigmoid(output)

  output = np.dot(output, W3) + b3
  output = softmax(output)

  return output
```

```python
x_test, y_test = get_data()
print(x_test.shape)
print(y_test.shape)
```

```python
network = init_network()
```

```python
y_predict = predict(network, x_test)
```

```python
acc = 0
for i in range(len(y_predict)):
  if y_test[i] == np.argmax(y_predict[i]):
    acc += 1
print(acc / len(y_test))
```
